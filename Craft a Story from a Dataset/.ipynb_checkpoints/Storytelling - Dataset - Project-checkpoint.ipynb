```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Import data

death_rate_data = pd.read_csv('death_data.csv')
```

```python
# Preview Data

death_rate_data.info()
```

```python
death_rate_data.describe()
```

```python
#preview
death_rate_data.head(10)
```

```python
# Check for missing values
print(death_rate_data.isnull().sum())
```

```python
death_rate_data = death_rate_data.fillna(0)
print(death_rate_data.isnull().sum())
```

###### Plot each year separately

```python
data = death_rate_data

sns.countplot(x='Years', data=data, hue='Cause')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the trends for each cause of death
plt.figure(figsize=(14, 8))

for column in death_rate_data.columns:
    if column != 'Years':
        plt.plot(death_rate_data['Years'], death_rate_data[column], marker='o', linestyle='-', label=column)

plt.title('Death Rates by Cause (1999-2021)')
plt.xlabel('Year')
plt.ylabel('Death Rate per 100,000 People')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

data = pd.read_csv('death_data.csv', parse_dates=['Years'], index_col='Years')
```

```python
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(15, 9))

for cause in data.columns:
        ax.plot(data.index, data[cause], label=cause)
ax.set_xlabel('Years')
ax.set_ylabel('Death Rate per 100,000 People')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
data = data.fillna(0)

data.columns
```

```python
sns.set(style="whitegrid")

columns1 = ["Accidents", "Alzheimer's disease",
       'Chronic liver disease and cirrhosis',
       'Chronic lower respiratory diseases', 'COVID-19', 'Diabetes', 'Homicide', 'Hypertension', 'Influenza and pneumonia',
       'Nephritis', "Parkinson's disease", 'Septicemia', 'Stroke', 'Suicide']

columns2 = ['Cancer', 'Heart disease']
fig, ax = plt.subplots(2, 1, figsize=(15, 8))

for cause in data.columns:
    ax[0].plot(data.index, data[columns2], label=cause)
    ax[1].plot(data.index, data[columns1], label=cause)
ax[1].set_xlabel('Years')
ax[1].set_ylabel('Death Rate per 100,000 People')

plt.show()
```

```python

# Filter the data for the years 2020 and 2021
data_2018_2021 = death_rate_data[death_rate_data['Years'].isin([2018, 2019, 2020, 2021])]

# Convert the data to long format for seaborn
data_2018_2021_long = data_2018_2021.melt(id_vars='Years', var_name='Cause', value_name='Death Rate')

# Set the seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 10))
sns.barplot(data=data_2018_2021_long, x='Cause', y='Death Rate', hue='Years')

plt.title('Comparison of Death Rates by Cause (2019 to 2021)')
plt.xlabel('Cause of Death')
plt.ylabel('Death Rate (per 100,000 people)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

columns1 = ["Accidents", "Alzheimer's disease",
       'Chronic liver disease and cirrhosis',
       'Chronic lower respiratory diseases', 'COVID-19', 'Diabetes', 'Homicide', 'Hypertension', 'Influenza and pneumonia',
       'Nephritis', "Parkinson's disease", 'Septicemia', 'Stroke', 'Suicide']

columns2 = ['Cancer', 'Heart disease']
fig, ax = plt.subplots(2, 1, figsize=(15, 8))

for cause in data.columns:
    ax[0].plot(data.index, data[columns2], label=cause)
    ax[1].plot(data.index, data[columns1], label=cause)
ax[1].set_xlabel('Years')
ax[1].set_ylabel('Death Rate per 100,000 People')

plt.show()
```

```python

# Infectious diseases
infect_diseases = ['COVID-19', 'Influenza and pneumonia', 'Septicemia']

sns.set(style="whitegrid")


fig, ax = plt.subplots(figsize=(10, 6))

for disease in infect_diseases:
    ax.plot(death_rate_data['Years'], death_rate_data[disease], label=disease)

ax.set_xlabel('Years')
ax.set_ylabel('Death Rate per 100,000 People')
ax.set_title('Infections Diseases')
ax.legend(loc='upper left') 

plt.show()
```

```python

# Neurological Conditions
neurological = ["Parkinson's disease", 'Stroke', "Alzheimer's disease"]

sns.set(style="whitegrid")


fig, ax = plt.subplots(figsize=(10, 6))

for n in neurological:
    ax.plot(death_rate_data['Years'], death_rate_data[n], label=n)

ax.set_xlabel('Years')
ax.set_ylabel('Death Rate per 100,000 People')
ax.set_title('Neurological Conditions')
ax.legend(loc='upper left') 

plt.show()
```

```python
# Cardiovascular Conditions
cardio_con0 = ["Heart disease"]
cardio_con1 = ["Hypertension"]
sns.set(style="whitegrid")


fig, ax = plt.subplots(2, 1, figsize=(10, 6))

for c in cardio_con0:
    ax[0].plot(death_rate_data['Years'], death_rate_data[c], label=c)
for c in cardio_con1:
    ax[1].plot(death_rate_data['Years'], death_rate_data[c], label=c)

ax[1].set_xlabel('Years')
ax[0].set_ylabel('Death Rate per 100,000 People')
ax[1].set_ylabel('Death Rate per 100,000 People')
ax[0].set_title('Cardiovascular Conditions')
ax[0].legend(loc='upper left') 
ax[1].legend(loc='upper left')
plt.show()
```
