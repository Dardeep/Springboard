```python
data = death_rate_data

sns.countplot(x='Years', data=data, hue='Cause')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the trends for each cause of death
plt.figure(figsize=(14, 8))

for column in death_rate_data.columns:
    if column != 'Years':
        plt.plot(death_rate_data['Years'], death_rate_data[column], marker='o', linestyle='-', label=column)

plt.title('Death Rates by Cause (1999-2021)')
plt.xlabel('Year')
plt.ylabel('Death Rate per 100,000 People')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.savefig(f'All_Death_Plot.png')
plt.show()
```

```python

data = pd.read_csv('death_data.csv', parse_dates=['Years'], index_col='Years')
```

```python
sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(15, 9))

for cause in data.columns:
        ax.plot(data.index, data[cause], label=cause)
ax.set_xlabel('Years')
ax.set_ylabel('Death Rate per 100,000 People')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig(f'All_Death_Plot.png')
plt.show()
```

```python
data = data.fillna(0)

data.columns
```

```python
sns.set(style="whitegrid")

columns1 = ["Accidents", "Alzheimer's disease",
       'Chronic liver disease and cirrhosis',
       'Chronic lower respiratory diseases', 'COVID-19', 'Diabetes', 'Homicide', 'Hypertension', 'Influenza and pneumonia',
       'Nephritis', "Parkinson's disease", 'Septicemia', 'Stroke', 'Suicide']

columns2 = ['Cancer', 'Heart disease']
fig, ax = plt.subplots(2, 1, figsize=(15, 8))

for cause in data.columns:
    ax[0].plot(data.index, data[columns2], label=cause)
    ax[1].plot(data.index, data[columns1], label=cause)
ax[1].set_xlabel('Years')
ax[1].set_ylabel('Death Rate per 100,000 People')

plt.show()
```

```python

# Filter the data for the years 2020 and 2021
data_2018_2021 = death_rate_data[death_rate_data['Years'].isin([2018, 2019, 2020, 2021])]

# Convert the data to long format for seaborn
data_2018_2021_long = data_2018_2021.melt(id_vars='Years', var_name='Cause', value_name='Death Rate')

# Set the seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 10))
sns.barplot(data=data_2018_2021_long, x='Cause', y='Death Rate', hue='Years')

plt.title('Comparison of Death Rates by Cause (2019 to 2021)')
plt.xlabel('Cause of Death')
plt.ylabel('Death Rate (per 100,000 people)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# Define the columns
cols = death_rate_data.columns

# Data for 2019, 2020, and 2021
death_2019 = death_rate_data.loc[death_rate_data['Years'] == 2019].set_index('Years')
death_2020 = death_rate_data.loc[death_rate_data['Years'] == 2020].set_index('Years')
death_2021 = death_rate_data.loc[death_rate_data['Years'] == 2021].set_index('Years')


for cause in cols:
    if cause != 'Years':
        # Percentage change
        percentage_change_2020 = ((death_2020[cause].values[0] - death_2019[cause].values[0]) / death_2019[cause].values[0]) * 100
        percentage_change_2021 = ((death_2021[cause].values[0] - death_2020[cause].values[0]) / death_2020[cause].values[0]) * 100

        # Arrows and Color
        arrow_2020 = '' if percentage_change_2020 >= 0 else ''
        color_2020 = 'green' if percentage_change_2020 >= 0 else 'red'
        
        # Determine the arrow direction and color for 2021
        arrow_2021 = '' if percentage_change_2021 >= 0 else ''
        color_2021 = 'green' if percentage_change_2021 >= 0 else 'red'

        # Plot
        fig, ax = plt.subplots(figsize=(8, 4))

        ax.text(0.5, 0.75, f'{int(death_2020[cause].values[0]):,}', horizontalalignment='center', verticalalignment='center', fontsize=40, fontweight='bold')
        ax.text(0.5, 0.50, f'{arrow_2020}{percentage_change_2020:.0f}%\n', horizontalalignment='center', verticalalignment='center', fontsize=17, color=color_2020)
        ax.text(0.75, 0.535, f'Change from 2019 to 2020', horizontalalignment='center', verticalalignment='center', fontsize=12)
        ax.text(0.5, 0.28, f'{arrow_2021}{percentage_change_2021:.0f}%\n', horizontalalignment='center', verticalalignment='center', fontsize=17, color=color_2021)
        ax.text(0.75, 0.315, f'Change from 2020 to 2021', horizontalalignment='center', verticalalignment='center', fontsize=12)
        ax.text(0.5, 0.9, cause, horizontalalignment='center', verticalalignment='center', fontsize=15, fontweight='bold')
        ax.axis('off')

        # Export Plot
        plt.savefig(f'{cause}_death_rate.png')

        
        plt.show()
```
