# Feature Engineering

Because the correlation isn't as strong it would be a good idea to combine or create new features that could ehance the performace of our ML model. 

Based on my understanding when it comes to financial loan underwriting, there are a few key indicators you have to consider when evaluating the strength of a borrower/applicant. 

1) Past credit history, in this case past 2 year delinquencies and public records. Combining the two may have higher risk 
We will create new feature: delinq_pub.rec (delinq.2yrs * pub.rec)

2) Income and cost of the loan. Combine log.annual.inc and installment amount. This will let us know the affordability of the loan relative to the borrwer's income. - income_loan_ratio (log.annual.inc / installment)

3) We will also do it in reverse, monthly payment to income ratio: installment_income_ratio (installment / log.annual.inc

4) debt to income with interest rate. This would get the debt load and the cost of their loan - dti_int.rate (dti * int.rate)

5) Total Credit utilization. Estimating the total credit limit available to the borrwer based on their revolving balance and utilization. total_credit_limit (revol.bal / (revol.util / 100)

```python

# Create a back up incase we mess this up. Data will be dffe (dataframe feature engineering)

dffe = df.copy()

# Create new features
dffe['delinq_pub.rec'] = dffe['delinq.2yrs'] * dffe['pub.rec']
dffe['income_loan_ratio'] = dffe['log.annual.inc'] / dffe['installment']
dffe['installment_income_ratio'] = dffe['installment'] / dffe['log.annual.inc']
dffe['dti_int.rate'] = dffe['dti'] * dffe['int.rate']
dffe['total_credit_limit'] = dffe['revol.bal'] / (dffe['revol.util'] / 100)

#handle infinite value error that I get later on when trying to use scaller and limit values 
dffe.replace([np.inf, -np.inf], np.nan, inplace=True)
dffe.fillna(0, inplace=True)

dffe.head()
```
