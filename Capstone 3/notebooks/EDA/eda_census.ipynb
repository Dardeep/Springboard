# Exploratory Data Analysis Cities Census Data

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
import numpy as np
import os
```

```python
census_df = pd.read_csv('../../data/census_df_sorted.csv')
census_df.head(30)
```

##### Summary

```python
summary_stats = census_df.describe()
print(summary_stats)
```

### Data Visualization

##### Distribution of 1-unit homes

```python
plt.figure(figsize=(10, 6))
sns.histplot(census_df['Total_1_Unit'], kde=True)
plt.title('Distribution of Total 1-Unit Homes Across Cities')
plt.xlabel('Number of 1-Unit Homes')
plt.ylabel('Frequency')
plt.show()
```

##### Top Cities with Total 1-unit Homes

```python
top_cities = census_df.sort_values(by='Total_1_Unit', ascending=False).head(25)

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Total_1_Unit', data=top_cities)
plt.title('Top 25 Cities by Total 1-Unit Homes')
plt.xlabel('City')
plt.ylabel('Total 1-Unit Homes')
plt.xticks(rotation=45)
plt.show()
```

### Correlation 
###### For the most part we don't need to find a correlation with this data, but it would be interesting to see how population, income and home values relate to the total number of 1-unit homes. Also to see if income and home values have a correlation. 

```python
numeric_df = census_df.select_dtypes(include=[float, int])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### Notes

There is an obvious correlation between total population and total 1 unit homes as larger population requires more housing unites. 

Income and Median Home value has a strong positive correlation. This suggests cities where residences earn more the home values are higher. 

Home value and 1unit homes, The weak positive correlation between home value and 1 unit homes suggests that home values do not influence the number of 1 unit homes in a city. 

1 unit Detached and 1 unit attached have a moderate correlation. Cities with more detached homes also tend to have more attached homes, this is mainly due to highly populated cities.
